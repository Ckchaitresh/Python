```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
a=np.linspace(1,10,5)
b=a*2
```

```python
plt.plot(a,b)
```

```python
plt.plot(a,b)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(a,b,'r')

plt.subplot(1,2,2)
plt.plot(b,a,'b')
```

```python
fig=plt.figure() # figure object created it is imaginary canvas we can add canvas to it
axes=fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
fig=plt.figure() # figure object created it is imaginary canvas we can add canvas to it
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(a,b,color='red',linestyle='--',marker='d') 
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Line Plot")
```

```python
fig=plt.figure() # figure object created it is imaginary canvas we can add canvas to it
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.bar(a,b,color='blue') 
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Line Plot")
```

```python
fig=plt.figure() # figure object created it is imaginary canvas we can add canvas to it
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.hist(a,b,color='green') 
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Line Plot")
```

```python
cities = ['Bangalore', 'Mysore', 'Mangalore', 'Hubli', 'Belgaum', 'Davanagere']
populations = [8410000, 920000, 625000, 1000000, 500000, 450000]
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,6))

axes[0].set_title('Population of Cities in Karnataka')
axes[0].set_xlabel('Cities')
axes[0].set_ylabel('Population')
axes[0].plot(cities, populations, color='b', linestyle='--')

axes[1].set_title('Population of Cities in Karnataka')
axes[1].set_xlabel('Cities')
axes[1].set_ylabel('Population')
axes[1].bar(cities, populations, color='g')
```

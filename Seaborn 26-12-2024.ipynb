```python
%matplotlib inline
```

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips') #dataset called tips which is supported by seaborn will be loaded
```

```python
tips.head()
```

```python
sns.histplot(tips['total_bill'],kde=True,bins=30) 
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(tips) #only numerical values
```

```python
sns.pairplot(tips,hue='sex') #hue will take categorical values that is for example sex=male and female
```

```python
sns.pairplot(tips,hue='smoker',palette='coolwarm')
```

```python
tips.head()
```

```python
sns.barplot(x=tips['sex'],y=tips['total_bill'],data=tips)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.stripplot(x='sex',y='total_bill',data=tips,jitter=True)
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```
